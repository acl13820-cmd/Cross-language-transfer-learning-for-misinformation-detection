```python
import numpy as np
import pandas as pd
import pickle
```

```python
en = pd.DataFrame(pd.read_csv('Data\WELFake_Dataset.csv'))
en=en[['title', 'text', 'label']]

en['label'] = en['label'].replace(0, 2)
en['label'] = en['label'].replace(1, 0)
en['label'] = en['label'].replace(2, 1)

en.head()

# Switch the numbers so now 1 = fake, 0 = real
```

```python
# merge title and text

def merge_title_content(df, title, content):
    df[title] = df[title].fillna('')
    df[content] = df[content].fillna('')
    df['merged_text'] =  df[title] + " " + df[content]
```

```python
merge_title_content(en, 'title', 'text')
en_merged = en[['label', 'merged_text']]
en_merged.head()
```

```python
cn = pd.DataFrame(pd.read_csv('Data\mcfend.csv'))
cn['label'] = cn['label'].replace('谣言', 1)
cn['label'] = cn['label'].replace('事实', 0)
cn = cn[cn['label'] != '尚无定论']

# drop unnecessary columns
cn = cn[['label', 'title', 'content']]

merge_title_content(cn, 'title', 'content')
cn_merged = cn[['label', 'merged_text']]
cn_merged.head()
```

```python
jp1 = pd.read_csv('Data\jpfake_news.csv')
jp1 = jp1[['context','isfake']]

# 0 = original/human-made
# 1 = partially generated
# 2 = completely generated 
# for the sake of this research I will presume all human-made news are factually correct, and i will replace all 1's and 2's as fake 

jp1['isfake'] = jp1['isfake'].replace([1,2],1)


jp1 = jp1.rename(columns = {'context' : 'merged_text', 
                      'isfake' : 'label'})
jp1.head()
```

```python
jp1['label'].value_counts()
```

```python
## IMPORTANT - the above JP dataset has only one column for context (hence no column for title) might have to consider combining title and content for other language resources
```

```python
jp2_label = pd.read_csv('Data\J-fakenews_v1\Label\label.tsv', sep='\t')
jp2_label.head()
```

```python
jp2_label['Q1'].unique()
```

```python
jp2_label['Q1'].value_counts()
```

```python
# for the sake of this research I will drop the suspended-judgment, unknown evidence, true category, treat the rest as misinformation. 

jp2 = jp2_label[['Article', 'Q1']]
jp2 = jp2[~jp2['Q1'].isin(['Unknown-Evidence', 'Suspended-Judgment'])]

jp2['Q1'] = jp2['Q1'].replace(['False', 'Misleading', 'Pants-on-Fire', 'False', 'Half-True', 'Inaccurate'], 1)
jp2['Q1'] = jp2['Q1'].replace('True', 0)

jp2 = jp2.rename(columns = {'Article' : 'merged_text', 
                      'Q1' : 'label'})
jp2['merged_text'] = jp2['merged_text'].str.replace('[FactCheck]', '')

jp2.head()
```

```python
jp = pd.concat([jp1, jp2], ignore_index=True)
jp.head()
```

```python
en.to_csv('en.csv', index = False)
cn.to_csv('cn.csv', index = False)
jp.to_csv('jp.csv', index = False)
```
